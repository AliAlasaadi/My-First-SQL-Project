```python
#imported the rquired libraries and classes to successfuly complete the project
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine,text
```

```python
#read the fist dataset named Cars containing the cars related details
df1=pd.read_csv('Cars.csv')
df1.head()
```

```python
#read the second dataset named Customers containing the customers information
df2=pd.read_csv('Customers.csv')
df2.head()
```

```python
#read the third dataset named Dealers contaning the dealers information
df3=pd.read_csv('Dealers.csv')
df3.head()
```

```python
#created a database named Car_sales.db using sqlite
engine=create_engine('sqlite:///Car_sales.db')
```

```python
#Assigned the dataframes of the datasets to tables
df1.to_sql('Cars_Table',engine, if_exists='replace')
df2.to_sql('Customers_Table',engine, if_exists='replace')
df3.to_sql('Dealers_Table',engine, if_exists='replace')
```

```python
#Selected all the data from the Cars_Table
query='Select * from Cars_Table'
result=pd.read_sql_query(query,engine)
result.head()
```

```python
#select all the data for the Ford Expedition cars
query='select * from Cars_table where Company="Ford" And Model ="Expedition"'
result= pd.read_sql_query(query,engine)
result.head()
```

```python
#select all the data based on the date of the sale from first to last
query='select * from Cars_Table order by date ASC'
result=pd.read_sql_query(query,engine)
result.head()
```

```python
#select only The company's name
query='select distinct Company from Cars_Table'
result=pd.read_sql_query(query,engine)
result.head()
```

```python
# change the the coulmn name'Price($)' To price
query = text('ALTER TABLE Cars_Table RENAME COLUMN "Price ($)" TO Price')

with engine.connect() as connection:
    connection.execute(query)

print("Column renamed successfully.")
```

```python
#Slect all the data to make sure the name was successfuly renamed
query='select * from Cars_Table'
result=pd.read_sql_query(query,engine)
result.head()
```
